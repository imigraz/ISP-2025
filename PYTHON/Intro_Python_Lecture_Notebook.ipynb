```python
# Running your first Python cell
print("Hello, World!")
```

```python
# Variables and Data Types
x = 10  # Integer
y = 3.14  # Float
name = "Python"  # String
is_learning = True  # Boolean
print(f"x: {x}, y: {y}, name: {name}, is_learning: {is_learning}")
```

```python
# Simple I/O
user_name = input("What's your name? ")
print(f"Welcome, {user_name}!")
```

```python
# Control Structures and Logic
number = int(input("Enter a number: "))
if number % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python
# Loop example: Iterating over a list
tasks = ["Learn Python", "Write code", "Test code"]
for task in tasks:
    print(f"Task: {task}")
```

```python
# Working with Lists
data = [5, 10, 15, 20]
data.append(25)
print(f"Updated list: {data}")

# Calculate the average of a list of numbers
def calculate_average(numbers):
    return sum(numbers) / len(numbers)

print(f"Average: {calculate_average(data)}")
```

```python
# Error Handling
try:
    number = int(input("Enter an integer: "))
    print(f"You entered: {number}")
except ValueError:
    print("That's not a valid integer!")
```

```python
# File Input/Output
# Writing to a file
with open("example.txt", "w") as file:
    file.write("Hello, Python!\n")
    file.write("This is an example file.")

# Reading from a file
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
# Using os and numpy
import os
import numpy as np

# List files in the current directory
print("Files in directory:", os.listdir("."))

# Numpy example: Create an array and perform calculations
array = np.array([1, 2, 3, 4, 5])
print(f"Array: {array}, Mean: {np.mean(array)}")
```

```python
# Pandas Data Analysis
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Score": [85, 90, 95]
}
df = pd.DataFrame(data)
print(df)

# Filter rows
filtered_df = df[df["Age"] > 25]
print(filtered_df)
```

```python
# Matplotlib Visualization
import matplotlib.pyplot as plt

# Basic plot
plt.plot(df["Name"], df["Score"], marker="o")
plt.title("Scores by Name")
plt.xlabel("Name")
plt.ylabel("Score")
plt.show()
```
